This is a very basic login and sign-up page

```python
def details():
  flag=False
  username=input("Hey!Enter your username: ")
  password=input("Enter your password: ")
  file1=open("file1.txt","r+")
  for line in file1:
    user,passw=line.split()
    if passw==password and user==username:
      flag=True
      break
  if flag:
    print(f"Hey! Welcome back {username}")
  else:
    print("We have signed you up!!!")
    signup(username,password)
```

The above function allows inputting of the credentials and checks if they are already present in the text file or not. The function prints "Hey! Welcome back" along with the username if the credentials are present when details() is called, otherwise  signup() is called.

```python
def signup(username,password):
  file1=open("file1.txt","a")
  file1.write(username)
  file1.write(' ')
  file1.write(password)
  file1.write("\n")
  file1.close()
```

The above function writes the username and password into the file with a space to differentiate them.

Now we finally call details() to execute the program.
